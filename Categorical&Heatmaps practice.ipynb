# Categorical Plots

```python
tips = sns.load_dataset("tips")
```

```python
tips.head()
```

```python
#barplot
sns.barplot(x="sex",y="total_bill",data=tips)
```

```python
sns.barplot(x="sex",y="total_bill",data=tips,estimator=np.std
#estimator shows output on the basis of function given            
```

```python
#countplot
sns.countplot(x="sex",data=tips)
#counts the number of occurrences
```

```python
#boxplots
#facilitates comparison bw variables

sns.boxplot(x="day",y="total_bill",data=tips)
#concept of whiskers and outliers
```

```python
sns.boxplot(x="day",y="total_bill",data=tips, hue="smoker")
```

```python
#violinplot
sns.violinplot(x="day",y="total_bill",data=tips)
#violinplot allows us to plot all the data values unlike the box plots
#little harder to interpret than boxplot
#more informative than boxplot
```

```python
sns.violinplot(x="day",y="total_bill",data=tips, hue="sex", split=True)
```

```python
#stripplot
sns.stripplot(x="day",y="total_bill",data=tips,jitter=False)
```

```python
sns.stripplot(x="day",y="total_bill",data=tips,jitter=True)
#jitter adds a bit of random noise so as to vizualise more clearly the datapoints
```

```python
sns.stripplot(x="day",y="total_bill",data=tips,jitter=True, hue="sex", dodge=True)
#split has been renamed to dodge
```

```python
sns.swarmplot(x="day",y="total_bill",data=tips)
#combination of violin and stripplot
```

```python
sns.catplot(x="day",y="total_bill",data=tips,kind="bar")
```

Matrix Plots(HEatmaps)
Matrix plots allow you to plot data as color-encoded matrices and can also be used to indicate clusters within the data

```python
flights=sns.load_dataset("flights")
flights.head()
```

```python
tc=tips.corr()
tc
```

```python
sns.heatmap(tc)
```

```python
sns.heatmap(tc, annot=True, cmap="coolwarm")
```

```python
flights
```

```python
#change it into matrix form: pivot_table
fl=flights.pivot_table(index="month",columns="year",values="passengers")
fl
```

```python
sns.heatmap(fl, linecolor="white", linewidth=1)
```

```python
sns.clustermap(fl)
#diff rows and columns are grouped together to show similarity bw them
```
