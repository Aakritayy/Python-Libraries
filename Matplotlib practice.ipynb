```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x=np.linspace(0,5,11)
x
```

```python
y = x**2
y
```

```python
#functional approach
plt.plot(x,y)
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("title")
```

```python
#creating multiplots
#using subplots

plt.subplot(1,2,1)
plt.plot(x,y,'r')
#specified r=  red as the color
plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
#OOP approach

fig=plt.figure()
#creating empty canvas

axes = fig.add_axes([0.1,0.1,0.8,0.8]) #left, bottom, width, height(ranging from 0 to 1)
axes.plot(x,y)
axes.set_xlabel("x axis")
axes.set_ylabel("y axis")
axes.set_title("title")
```

```python
fig=plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2= fig.add_axes([0.2,0.5,0.4,0.3])
axes1.plot(x,y,'b')
axes2.plot(y,x,'g')
```

```python
fig, axes = plt.subplots(nrows=1,ncols=2)

axes[0].plot(x,y)
axes[1].plot(y,x)
```

```python
axes
#array type
```

```python
fig, axes = plt.subplots(nrows=3,ncols=3)
plt.tight_layout
#gives spacing bw layouts
```

# FIGSIZE AND DPI(DOTS PER INCH OR PIXELS PER INCH)

```python
fig = plt.figure(figsize=(8,2))

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,'b')
```

```python
fig, axes = plt.subplots(1,2,figsize=(4,4))

axes[0].plot(x,y)
axes[1].plot(y,x)
```

```python
fig = plt.plot(figsize=(4,4))
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y**2)
axes.plot(y,x**3)
```

#plot appearance

```python
fig = plt.figure()

ax= fig.add_axes([0,0,1,1])
ax.plot(x,y, color='#FF57C0')
#ax.plot(x,y, color='orange')
#color parameter accepts basic color names, rgb hex forms

ax.plot(x,y, color='#FF57C0', lw=0.5,ls='--', marker='o', markersize=10,
       markerfacecolor="yellow", markeredgewidth=3, markeredgecolor="green")
#lw= linewidth
#alpha= controls the transparency of line
#ls = linestyle, "--", "-.", ":"
#marker: marks y vs x, "o", "+", "*" symbols can be used
```
